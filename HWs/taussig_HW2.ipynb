```python
cols = ['Gender','Age','Race1','Education','HHIncome','Weight','Height','BMI',
        'Pulse','TotChol','AlcoholYear','Smoke100','RegularMarij','HardDrugs',
        'Diabetes','Depressed']

trimmed = data[cols]
clean = trimmed.dropna()

clean.shape[0]
```

BMI and Weight seem to have the best fit because BMI is based off height and weight. So the variables are based on each other to deteermine BMI. For some comparions like BMI and education, a lot of the time these things don't have to do with each other as much as weight and bmi do.

```python
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt

plt.style.use('dark_background')
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
import pandas as pd

plt.style.use('dark_background')

Y = clean['BMI'].astype(float).values

for col in clean.columns:
    if col == 'BMI':
        continue

    X_series = clean[col]

    if not pd.api.types.is_numeric_dtype(X_series):
        X_series = X_series.astype('category').cat.codes

    X = X_series.astype(float).values.reshape(-1, 1)

    lin = LinearRegression().fit(X, Y)
    Yhat = lin.predict(X)

    order = np.argsort(X[:, 0])
    Xs = X[order, 0]
    Ys = Y[order]
    Yhs = Yhat[order]

    plt.figure(figsize=(10,8))
    plt.scatter(Xs, Ys, color='yellow', zorder=5)
    plt.plot(Xs, Yhs, color='green', zorder=3)
    plt.vlines(Xs,
               ymin=np.minimum(Ys, Yhs),
               ymax=np.maximum(Ys, Yhs),
               color='red')
    plt.xlabel(col)
    plt.ylabel('BMI')
    plt.title(f'BMI vs {col}')
    plt.show()
```
