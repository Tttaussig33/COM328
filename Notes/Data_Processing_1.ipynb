We start our week by learning three core concepts (I)boolean expressions/comparisons, (II)conditionals and finally (III)looping:
Ia. Boolean expression are foundational and form the inner logical core to more complex program elements. Very simply we have a boolean type which is True or False.  Next we have a few operators that allow us to form conjunctive and disjunctive expressions (a)  X or Y : True if either X or Y is True, (b) X and Y : True only if X is True and Y is True, (c) not X : True if X is False.

Ib. Comparison operators are the foundation of more complex program elements such as finding the minimum element, sorting and indexing a B+tree in a database (for a primary key).  They are binary in form (like or,and) (a) X < 1 : less than 1, (b) X <= 1 : less than or equal to 1, (c) X==1 : equal to 1, (d) X!=1 : not equal to 1, (e) X>=1 greater than or equal to 1, (f) X>1 greater than 1.  Notice that we could express (c) X==1 by: (not (X<1)) and (not (X>1)).  As a direct result you will find boolean expressions and comparisons as solutions to problems often can take many correct forms but have a single compact form (IE there is a way to reduce a series of boolean expressions to a minimum set thereby running faster…)

II. Conditional expressions are foundational structures that allow you to run a program block on the condition of the expression.  In python you have only one main way to do this:
if condition :
    code block that will run when if condition is True
else:
    code block that will run when condition is False

```python
#[Example: condition]
x = 0
if x<1:
    print('Hello')
else:
    print('World')
```

more at the tutorial is here: https://docs.python.org/3/tutorial/controlflow.html#if-statements

III. There are three pure python ways to loop: (a) while (b) for (c) recursive function
We will focus on using the while and for loop and save (c) for those that are interested/advanced and wish to be able to work with complex, non-linear data structures such as trees and graphs.
(a) while loop (more flexible in general, but requires managing  a control variable)
while condition:
    code block that will be run as long as the condition is True
    usually you will then take some action that will get you closure to ending the loop

```python
#[Example: while]
x = 0
while x<10:
    print('Hello')
    x += 1
```

When we forget to update the counting variable x in this example the loop becomes infinite (a common mistake when using the while)
(b) for loop (easy for traversing lists, dicts, arrays)
for count_iterator:
    code block that will be run as long as the count iterator keeps moving

```python
#[Example: for 1]
for x in range(10):
    print('Hello')

#[Example: for 2]
X = [i for i in range(10)]
```

Notice that this is the same as the while example and is easier because it is a fixed number of times that we want to loop.  When you have an idea of how many times you want to do something, the for loop often is the better choice.  The while is best saved for special uses since it requires managing the counting or sentimental variable separately (IE more that can go wrong). Example 2 shows a special type of loop in python called list comprehension.  This form is faster (more on this later).
