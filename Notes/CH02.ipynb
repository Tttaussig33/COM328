### Overview

#### Linear Algebra Preliminaries (Datasets, Variables, Rows, Columns)

Whether we have prior datasets that we are try to fit a model to, or a new unseen dataset, we can consider either of them as a matrix, where the number of rows is $n$ and the number of dimensions (variables, columns) is $p$:

$X = 
\begin{pmatrix}
x_{11} & x_{12} & \cdots & x_{1p} \\
x_{21} & x_{22} & \cdots & x_{2p} \\
\vdots  & \vdots  & \ddots & \vdots  \\
x_{n1} & x_{n2} & \cdots & x_{np} 
\end{pmatrix}$

Traditionally they are homogeneous in nature but we will relax that specific requirement as we explore the differences between quantitative and qualitative columns vectors (variables or fields). If we want the $i$th row vector (data point or observation) we write is as:

$x_i =
\begin{pmatrix}
  x_{i1} \\
  x_{i2} \\
  \vdots \\
  x_{ip}
 \end{pmatrix}
$ $\;\;\;$ Note: in LA it is not the transpose: $x_i^{T} =
\begin{pmatrix}
  x_{i1} && x_{i2} \cdots && x_{ip}
 \end{pmatrix}
$


And if we want the $j$th column vector (variable or field) we would have:

$x_j =
\begin{pmatrix}
  x_{1j} \\
  x_{2j} \\
  \vdots \\
  x_{nj}
 \end{pmatrix}
$

So vectors (column or row) are vertical, while the transpose of the vector is horizontal.  Finally we can rotate a matrix causing columns to be rows by using the matrix transpose operation:

$X^{T} = 
\begin{pmatrix}
x_{11} & x_{21} & \cdots & x_{n1} \\
x_{12} & x_{22} & \cdots & x_{n2} \\
\vdots  & \vdots  & \ddots & \vdots  \\
x_{1p} & x_{2p} & \cdots & x_{np} 
\end{pmatrix}$

#### Predictor Framework

If we have a dataset now of quantitative values called matrix $A$ we will usually partition column vectors into a single response vector and then $p$ remaining predictor vectors: $X_1,X_2, \cdot\cdot\cdot ,X_p$. We are trying to find some kind of explainable relationship between $Y$ and $X = (X_1,X_2, \cdot\cdot\cdot ,X_p)$ so we often start with a very simple framework to explore:

$Y = f(X) + \epsilon$ 

$\epsilon$ is an error term that we assume to have no loading, trend or drift (zero-mean centered...)

Statistical learning then is the process of trying to guess $f$.  In contrast, Machine Learning is not very concerned with $f$ but are mostly interested in the accuracy of the estimate (how good is it?) and then in turn given high-accuracy the actual values that are predicted by the ML models.
