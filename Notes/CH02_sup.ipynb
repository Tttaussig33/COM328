#### <u>Sample Mean (Expected Value)<u>
We will first start with the mean of any of our $x_j$s, where $\hat{X}$ is the sample mean and $E(x_j)$ is the expected value of $x_j$:

#### $E(x_j) = \hat{x_j} = \left(\frac{1}{n}\right)\displaystyle\sum_{i=1}^{n}x_{ij}$

If we look at the first column vector then $j=1$ we have:

$E(x_1) = \hat{x_j} = \left(\frac{1}{7}\right)\displaystyle\sum_{i=1}^{7}x_{ij} = \left(\frac{1}{7}\right)(1+2+4+1+2+3+4) = \frac{17}{7}$

#### <u>Sample Variance/Sample Standard Deviation<u>

#### $Var(x_j) = s^2 = \left(\frac{1}{n-1}\right)\displaystyle\sum_{i=1}^{n}(x_{ij}-E(x_j))$

$s = \sqrt{s^2}$ : sample standard deviation is just the square root of the variance
    
If we look at the first column vector then $j=1$ we have:

$Var(x_1) = \left(\frac{1}{7-1}\right)\displaystyle\sum_{i=1}^{7}[x_{i1}-E(x_1)]^2 = \left(\frac{1}{6}\right)((1-1/7)^2+(2-1/7)^2+(4-1/7)^2+(1-1/7)^2+(2-1/7)^2+(3-1/7)^2+(4-1/7)^2) = 1.61...$

#### <u>Correlation<u>

Pearson correlation coeefficent is a linear relationship test bewteen two column vectors of a dataset with $n$ row vectors (observations).  It is defined for column vectors $x$ and $y$ as:
    
$r_{xy} = \displaystyle\frac{\sum_{i=1}^{n} (x_i-\bar{x})(y_i-\bar{y}) }{\sqrt{ \sum_{i=1}^{n}(x_i-\bar{x})^2 \sum_{i=1}^{n}(y_i-\bar{y})^2}} = 
\frac{\sum_{i=1}^{n} (x_i-\bar{x})(y_i-\bar{y}) }{(n-1) \sqrt{ s_x^2 s_y^2 }}
$

Here will will look to see if our datset column vectors $x_1, x_2$ are correlated by calculating parts of the above definition:


```python
n    = len(x_1)
x    = x_1
xbar = np.sum(x)/n
s2x  = np.sum((x-xbar)**2)/(n-1)
y    = x_2
ybar = np.sum(y)/n
s2y  = np.sum((y-ybar)**2/(n-1))
np.sum((x-xbar)*(y-ybar)/((n-1)*np.sqrt(s2x*s2y))),np.corrcoef(x,y)[0,1]
```
